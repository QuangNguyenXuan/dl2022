```python
# 1x1 conv
import tensorflow as tf
from tensorflow import keras


height  = 224
width   = 224
channel = 3

# model
model = keras.models.Sequential()
model.add(tf.keras.Input(shape=(height, width, channel)))

# block-1
model.add(keras.layers.Conv2D(64, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(64, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-2
model.add(keras.layers.Conv2D(256, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(256, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-3
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-4
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-5
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# Dense block
model.add(keras.layers.Conv2D(4096, 7, activation='relu'))
model.add(keras.layers.Conv2D(4096, 1, activation='relu'))
model.add(keras.layers.Conv2D(10, 1, activation='softmax'))

model.summary()
```

```python
# fully connected
import tensorflow as tf
from tensorflow import keras


height  = 224
width   = 224
channel = 3

# model
model = keras.models.Sequential()
model.add(tf.keras.Input(shape=(height, width, channel)))

# block-1
model.add(keras.layers.Conv2D(64, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(64, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-2
model.add(keras.layers.Conv2D(256, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(256, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-3
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-4
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# block-5
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.Conv2D(512, 3, padding='same', activation='relu'))
model.add(keras.layers.MaxPooling2D(2))

# Dense block
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(4096, activation='relu'))
model.add(keras.layers.Dense(4096, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

model.summary()
```

```python
# fully connected - loop
import tensorflow as tf
from tensorflow import keras

# model
model = keras.models.Sequential()
model.add(tf.keras.Input(shape=(32, 32, 3)))

num_of_blocks = 5
convs   = [2, 2, 3, 3, 3]
filters = [64, 256, 512, 512, 512]

# 5 blocks
for block in range(num_of_blocks):
    for conv in range(convs[block]):
        model.add(keras.layers.Conv2D(filters[block], 3, 
                                      padding='same', 
                                      activation='relu'))
    model.add(keras.layers.MaxPooling2D(2))

# Dense blocks
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

model.summary()
```
