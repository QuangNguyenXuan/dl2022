```python
# fashion_mnist
import tensorflow as tf
from tensorflow import keras

# Data Preparation
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print(type(y_test))
```

```python
import matplotlib.pyplot as plt 
import numpy as np

# Tạo dang sách 9 phần tử ngẫu nhiên từ m_train có 60000 phần tử
indices = list(np.random.randint(X_train.shape[0], size=64))

fig = plt.figure(figsize=(9,9))
columns = 8
rows = 8
for i in range(1, columns*rows +1):
    img = X_train[indices[i-1]].reshape(28,28)
    fig.add_subplot(rows, columns, i)
    
    plt.axis('off')
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
# cifar10
import tensorflow as tf
from tensorflow import keras

# Data Preparation
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

fig =plt.figure(figsize=(12,12))

for i in range(10):
    X_train_i = X_train[y_train[:,0]==i]    
    X_train_i = X_train_i[:10]
    
    for j in range(10):
        img = X_train_i[j].reshape(32,32,3)
        fig.add_subplot(10, 10, i*10 + j+1)
        plt.axis('off')
        plt.imshow(img, vmin=0, vmax=255)
```

```python
from PIL import Image
import numpy as np

indices = list(np.random.randint(X_train.shape[0], size=10))
for i in range(10):
    im = Image.fromarray(X_train[indices[i]].reshape(32,32,3))
    im.save("images/image_" + str(i) +".png")  
```
