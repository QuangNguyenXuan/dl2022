```python
vgg16 = models.vgg16(pretrained=True)
f_extractor = vgg16.features

# Freeze the first five layers
layer_count = 0
for child in f_extractor.children():
    if layer_count < 5:
        for param in child.parameters():
            param.requires_grad = False
    layer_count += 1

# Check
for name, param in vgg16.named_parameters():
    print(f"{name}: {param.requires_grad}")
```

```python
import torch
import torchvision.models as models

# Load the pretrained VGG16 model
vgg16 = models.vgg16(pretrained=True)

# Modify the classifier
# The 'classifier[6]' is typically the last fully connected layer of VGG16.
# We replace it with a new Linear layer with 10 outputs (for CIFAR-10 classes).
vgg16.classifier[6] = torch.nn.Linear(vgg16.classifier[6].in_features, 10)

# Print the modified model to confirm changes
print(vgg16)
```
